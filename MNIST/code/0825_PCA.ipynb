```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.decomposition import PCA
```

```python
train = np.load('data/train.npy', allow_pickle = 'True')
test = np.load('data/test.npy', allow_pickle = 'True')
```

```python
train_image = train[:,2:].reshape(-1,28,28).astype(np.int)
train_digit = train[:,0]
train_letter = train[:,1]

test_image = test[:,1:].reshape(-1,28,28).astype(np.int)
```

```python
type(train_digit[0])
```

```python
train_0 = []
for idx in range(len(train)):
    if train_digit[idx] == 0:
        train_0.append(train_image[idx])
train_0 = np.array(train_0)
```

```python
train_0.shape
```

```python
idx = 0
plt.figure(figsize=(5,5))
plt.subplot(1,2,1)
plt.imshow(train_0[idx])
plt.axis('off')

#plt.title('digit:{}   letter:{}'.format(train_digit[idx], train_letter[idx]), 
#          loc='left', fontsize=15)

plt.subplot(1,2,2)
data = np.where(train_0>=140, train_0, 0)
plt.imshow(data[idx].reshape(28,28))
plt.axis('off')
```

```python
pca = PCA(n_components=5)
x = train_0.reshape(train_0.shape[0],-1).copy()
w = pca.fit_transform(x)
x_ = pca.inverse_transform(w)
```

```python
print(w.shape)
print(x_.shape)
```

```python
plt.figure(figsize=(3,3))
plt.imshow(pca.components_[0].reshape(28,28))
plt.axis('off')
```

```python
fig, axes = plt.subplots(1, 5,
                         subplot_kw={'xticks':(),'yticks':()})

for i, (comp, ax) in enumerate(zip(pca.components_, axes.ravel())):
    ax.imshow(comp.reshape(28,28))
    ax.set_title('principal component {}'.format(i+1))
plt.show()
```
