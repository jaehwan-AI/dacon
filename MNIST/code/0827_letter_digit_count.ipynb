```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.decomposition import PCA
```

```python
train = pd.read_csv('data/train.csv')
```

```python
letters_sorted = np.sort(train['letter'].unique())
for letter in letters_sorted:
    digits = train[train['letter'] == letter]['digit']
    count = digits.value_counts()
    tmp_df = pd.DataFrame({letter:count})
    if letter == 'A':
        df = tmp_df
    else:
        df = pd.concat([df, tmp_df], axis=1)
```
