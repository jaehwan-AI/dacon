```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.decomposition import PCA
```

```python
train = np.load('data/train.npy', allow_pickle = 'True')
test = np.load('data/test.npy', allow_pickle = 'True')
```

```python
train_image = train[:,2:].reshape(-1,28,28).astype(np.int)
train_digit = train[:,0]
train_letter = train[:,1]
train_del_digit = np.where(train_image<=90, train_image, 90)

test_image = test[:,1:].reshape(-1,28,28).astype(np.int)
```

```python
train_del_noise = np.where(train_image>=150, train_image, 0)

train_5 = []
for idx in range(len(train)):
    if train_digit[idx] == 5:
        train_5.append(train_del_noise[idx])
train_5 = np.array(train_5)

pca = PCA(n_components=5)
x = train_5.reshape(train_5.shape[0],-1).copy()
w = pca.fit_transform(x)
x_ = pca.inverse_transform(w)

fig, axes = plt.subplots(1, 5,
                         subplot_kw={'xticks':(),'yticks':()})

for i, (comp, ax) in enumerate(zip(pca.components_, axes.ravel())):
    ax.imshow(comp.reshape(28,28))
    ax.set_title('principal component {}'.format(i+1))
plt.show()
```

```python
p = pca.components_
pca_re = PCA(n_components=1)
p_w = pca_re.fit_transform(p)

figure_p = pca_re.components_.reshape(28,28)
plt.imshow(figure_p)
plt.axis('off')
```

```python
def PCA_letter(data, letter, num):
    lst = []
    for idx in range(len(data)):
        if train_letter[idx] == letter:
            lst.append(data[idx])
    lst = np.array(lst)
    
    pca = PCA(n_components=num)
    x = lst.reshape(lst.shape[0],-1).copy()
    w = pca.fit_transform(x)
    return pca.components_
```

```python
def PCA_letter_mean(data, letter, num):
    lst = []
    for idx in range(len(data)):
        if train_letter[idx] == letter:
            lst.append(data[idx])
    lst = np.array(lst)
    
    pca = PCA(n_components=num, whiten=True)
    x = lst.reshape(lst.shape[0],-1).copy()
    w = pca.fit_transform(x)
    return pca.components_, pca.mean_
```

```python
com_, mean_ = PCA_letter_mean(train_del_digit, 'B', 3)
plt.imshow(mean_.reshape(28,28))
plt.axis('off')
```
